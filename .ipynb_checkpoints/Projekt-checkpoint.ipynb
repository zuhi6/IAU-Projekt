```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
personal_data = pd.read_csv('./data/personal_train.csv')
other_data = pd.read_csv('./data/other_train.csv')
personal_data = personal_data.rename(columns = {'Unnamed: 0':'id'})
personal_data.count()['id']
other_data = other_data.rename(columns = {'Unnamed: 0':'id'})
```

```python
fig_dims = (2, 1)

plt.subplot2grid(fig_dims, (1, 0))
personal_data['sex'].value_counts().plot(kind='bar', 
                                    title='Gender Counts')
plt.xticks(rotation=0)
```

```python
personal_data[(personal_data['sex'] != 'F') & (personal_data['sex'] != 'M')].count()["id"]
```

```python
personal_data.head()
```

```python
male = personal_data[personal_data['sex'] == 'M']
male = male[~np.isnan(male['age'])]

female = personal_data[personal_data['sex'] == 'F']
female = female[~np.isnan(female['age'])]

data = [male['age'], female[female['age']< 150]['age'], personal_data[personal_data['age']<150]['age']]

fig1, ax1 = plt.subplots()
ax1.set_title('Box plot of Males, Females and General')
ax1.boxplot(data)
plt.xticks([1,2,3], ['Male', 'Female', 'Together'])
plt.show()
```
